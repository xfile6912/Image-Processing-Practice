```python
import numpy as np
import os
import cv2
from imageio import imread
import matplotlib.pyplot as plt
%matplotlib inline
print("PACKAGE LOADED")
```

```python
cwd=os.getcwd()
print("Current folder is %s" %(cwd))
```

```python
def print_typeshape(img):
    print("Type is %s" %(type(img)))
    print("Shape is %s" % (img.shape, ))
```

```python
cat = cv2.cvtColor(cv2.imread("./../img/cat.jpg"), cv2.COLOR_BGR2RGB);
print_typeshape(cat)
plt.figure(0)
plt.imshow(cat)
plt.title("CAT")
plt.draw()
```

```python
cat2=np.array(cat).astype(float)
print_typeshape(cat2)
plt.figure(0)
plt.imshow(cat2/255)
plt.title("Original Image with imread.astype(float)/255")
plt.draw()
```

```python
catsmall = cv2.resize(cat, dsize=(100, 100), interpolation=cv2.INTER_CUBIC)
print_typeshape(catsmall);

plt.figure(1)
plt.imshow(catsmall)
plt.title("resized image")
plt.draw()
```

```python
def rgb2gray(rgb):
    if len(rgb.shape) == 3:
        return np.dot(rgb[...,:3], [.299, 0.587, 0.114])
    else:
        print("Current Image if Gray!")
        return rgb
```

```python
catsmallgray = rgb2gray(catsmall)
print_typeshape(catsmallgray)

plt.imshow(catsmallgray, cmap=plt.get_cmap("gray"))
plt.title("[imshow] Gray Image")
plt.draw()
```

```python
#MATRIX TO VECTOR
catrowvec=np.reshape(catsmallgray, (1, -1))
print("size of catrowvec is" , catrowvec.shape);
print("type of catrowvec is", type(catrowvec));

#VECTOR TO MATRIX
catmatrix=np.reshape(catrowvec, (100, -1))
print("size of catmatrix is" , catmatrix.shape);
print("type of catmatrix is", type(catmatrix));
```

```python
plt.imshow(catmatrix, cmap=plt.get_cmap("gray"))
plt.title("Reshaped Matrix");
plt.colorbar();
plt.draw();
```

```python
cwd = os.getcwd()
path = cwd+"/../img/";
flist=os.listdir(path)
print("[%d] File are in [%s]" % (len(flist), path));
```

```python
for i, f in enumerate(flist):
    print("%dth file is %s" %(i, f))
```

```python
valid_exts = [".jpg", ".gif", ".png", ".tga", ".jpeg"]
imgs=[]
names=[]
for f in flist:
    ext=os.path.splitext(f)[1]
    if ext.lower() not in valid_exts:
        continue;
    #Load and append to imgs list
    fullpath=os.path.join(path, f)
    imgs.append(cv2.cvtColor(cv2.imread(fullpath), cv2.COLOR_BGR2RGB))
    #append names
    names.append(os.path.splitext(f)[0])
```

```python
for img, name in zip(imgs, names):
    plt.imshow(img)
    plt.title(name)
    plt.show()
```
